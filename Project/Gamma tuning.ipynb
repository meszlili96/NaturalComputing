```python
import os
import numpy as np
from glob import glob
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plot_results(path):
    # KDE 
    kde = cv2.imread(path+"/KDE test.png")
    kde = cv2.cvtColor(kde, cv2.COLOR_BGR2RGB) 
    
    plt.figure(figsize=(36,18))
    plt.imshow(kde)
    plt.show()
    
    jsd = np.load(path+"/jsd.npy")
    hq_rate = np.load(path+"/hq_rate.npy")
    x_stdev = np.load(path+"/x_stdev.npy")
    ll = np.load(path+"/fs_ll.npy")
    
    fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, figsize=(7,15))

    ax1.plot(jsd)
    ax1.axhline(y=0, color='tab:red')
    
    ax2.plot(hq_rate)
    ax2.axhline(y=1, color='tab:red')
    
    ax3.plot(x_stdev)
    ax3.axhline(y=0.02, color='tab:red')
    
    ax4.plot(ll)
    
    plt.show()
```

```python
folder = "Gamma tuning/8 gauss egan, gamma 0.0"
plot_results(folder)
```

## Comparison of the best results

```python
# Candidates: 0.0, 0.05, 0.1, 0.4
path1 = "Gamma tuning/8 gauss egan, gamma 0.0"
path2 = "Gamma tuning/8 gauss egan, gamma 0.05"
path3 = "Gamma tuning/8 gauss egan, gamma 0.1"
path4 = "Gamma tuning/8 gauss egan, gamma 0.4"
```

```python
# KDE:
kde1 = cv2.imread(path1+"/KDE test.png")
kde2 = cv2.imread(path2+"/KDE test.png")
kde3 = cv2.imread(path3+"/KDE test.png")
kde4 = cv2.imread(path4+"/KDE test.png")

fig, axis = plt.subplots(2, 2, figsize=(20,20))
axis[0, 0].imshow(kde1)
axis[0, 0].axis(False)
axis[0, 1].imshow(kde2)
axis[0, 1].axis(False)
axis[1, 0].imshow(kde3)
axis[1, 0].axis(False)
axis[1, 1].imshow(kde4)
axis[1, 1].axis(False)
plt.tight_layout()
plt.show()
```

```python
# JSD
jsd1 = np.load(path1+"/jsd.npy")
jsd2 = np.load(path2+"/jsd.npy")
jsd3 = np.load(path3+"/jsd.npy")
jsd4 = np.load(path4+"/jsd.npy")

plt.figure(figsize=(16,8))
plt.plot(jsd1, label="0.0")
plt.plot(jsd2, label="0.05")
plt.plot(jsd3, label="0.1")
plt.plot(jsd4, label="0.4")
plt.axhline(y=0, color='k')
plt.xlabel("Epoch")
plt.ylabel("JSD(nats)")
plt.legend()
plt.show()
```

```python
# x_stdev
x_stdev1 = np.load(path1+"/x_stdev.npy")
x_stdev2 = np.load(path2+"/x_stdev.npy")
x_stdev3 = np.load(path3+"/x_stdev.npy")
x_stdev4 = np.load(path4+"/x_stdev.npy")

plt.figure(figsize=(16,8))
plt.plot(x_stdev1, label="0.0")
plt.plot(x_stdev2, label="0.05")
plt.plot(x_stdev3, label="0.1")
plt.plot(x_stdev4, label="0.4")
plt.axhline(y=0.02, color='k')
plt.xlabel("Epoch")
plt.ylabel("Stdev x")
plt.legend()
plt.show()

print("Last 20 epochs X stdev range, gamma 0.0 min: {}, max: {}".format(x_stdev1[-20:].min(), x_stdev1[-20:].max()))
print("Last 20 epochs X stdev range, gamma 0.05 min: {}, max: {}".format(x_stdev2[-20:].min(), x_stdev2[-20:].max()))
print("Last 20 epochs X stdev range, gamma 0.1 min: {}, max: {}".format(x_stdev3[-20:].min(), x_stdev3[-20:].max()))
print("Last 20 epochs X stdev range, gamma 0.4 min: {}, max: {}".format(x_stdev4[-20:].min(), x_stdev4[-20:].max()))
```

```python
# HQ rate
hq_rate1 = np.load(path1+"/hq_rate.npy")
hq_rate2 = np.load(path2+"/hq_rate.npy")
hq_rate3 = np.load(path3+"/hq_rate.npy")
hq_rate4 = np.load(path4+"/hq_rate.npy")

plt.figure(figsize=(16,8))
plt.plot(hq_rate1, label="0.0")
plt.plot(hq_rate2, label="0.05")
plt.plot(hq_rate3, label="0.1")
plt.plot(hq_rate4, label="0.4")
plt.axhline(y=1, color='k')
plt.xlabel("Epoch")
plt.ylabel("High quality rate")
plt.legend()
plt.show()

print("High quality rate, gamma 0.0: {}".format(hq_rate1[-1]))
print("High quality rate, gamma 0.05: {}".format(hq_rate2[-1]))
print("High quality rate, gamma 0.1: {}".format(hq_rate3[-1]))
print("High quality rate, gamma 0.4: {}".format(hq_rate4[-1]))
```

```python
# HQ rate
ll1 = np.load(path1+"/fs_ll.npy")
ll2 = np.load(path2+"/fs_ll.npy")
ll3 = np.load(path3+"/fs_ll.npy")
ll4 = np.load(path4+"/fs_ll.npy")

plt.figure(figsize=(16,8))
plt.plot(ll1, label="0.0")
plt.plot(ll2, label="0.05")
plt.plot(ll3, label="0.1")
plt.plot(ll4, label="0.4")
plt.xlabel("Epoch")
plt.ylabel("Real sample log likelihood")
plt.legend()
plt.axis([-5, 105, -450, 20])
plt.show()
```
